```python
import nemo.collections.asr as nemo_asr
asr_model = nemo_asr.models.ASRModel.from_pretrained(model_name="nvidia/parakeet-tdt-0.6b-v2")
```

```python
import time
import wave
from pydub import AudioSegment

#Load the file path
mp3_path = "/content/audio.mp3"
wav_path = "/content/audio.wav"

sound = AudioSegment.from_file(mp3_path)
sound.export(wav_path, format="wav")

with wave.open(wav_path, "rb") as audio_file:
    frames = audio_file.getnframes()
    rate = audio_file.getframerate()
    duration = frames / float(rate)

print(f"Audio Duration: {duration:.2f} seconds")

#calculate latency
start_time = time.time()

output = output = asr_model.transcribe([wav_path])

end_time = time.time()
latency = end_time - start_time

# Calculate real time
rtf = latency / duration

# 5. Output results
print(f"Latency: {latency:.2f} seconds")
print(f"Real-Time Factor (RTF): {rtf:.2f}")
print(f"Transcription: {output[0].text}")
```
